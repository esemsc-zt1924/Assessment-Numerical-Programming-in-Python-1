# Useful definitions for the assessment.

This notebook gives useful information and specific definitions for some elements of the assessment. Note that the descriptions given here are not universal.

## Weather data

Weather data is recorded at many sites across the UK, both for forecasting purposes and in order to provide instant data (for example at airfields).

### Longitude

The longitude of the site, given in degrees. This is the east-west position of the site, with 0 degrees being the Prime Meridian in Greenwich. Nowadays this is usually given on the WGS84 datum to match the Global Positioning System data for the site.

### Latitude

The latitude of the site, given in degrees. This is the north-south position of the site, with 0 degrees being the equator. This is also usually given on the WGS84 datum.

### Air Temperature
Air temperature is temperature as we commonly use it, recorded in degrees Celcuius, in the shade (usually inside a weather station stored in white painted box behind a Stephenson screen) at a height of 1.5m above the ground.

### Dew Point

The dew point is the temperature at which air becomes fully saturated, and thus water vapour can begin to condense into precipitation. This is indirectly calculated from other measurements and is given in degrees Celcius.

### Mean Sea Level Pressure
The pressure which would be recorded if the site were at mean sea level. This is calculated based on the observed pressure and the height of the site above mean sea level, and is given in hPa.

## Relative Humidity

The percentage of water vapour in the air compared to the maximum it could contain at the current temperature and pressure. This is calculated from other variables.

### Wind Speed

The instantaneous mean wind speed at the site in m/s.

### Wind Direction

The wind direction in degrees. This is given as the direction the wind is coming from.

### Wind Gust

The 5% highest wind speed recorded over the validity time. This is in m/s.

### Visibility

Visibility on a ten point scale, corresponding to ranges of distances at which a specific object is identifiable from the background. Generally the lowest value (0) corresponds to a visibility in tens of meters, while the highest value corresponds to 40 km or more.

## The Rankine Temperature Scale

The Rankine temperature scale is an absolute temperature scale, similar to the Kelvin scale you may have already heard about. The Rankine scale is defined such that 0 degrees Rankine is absolute zero (-273.16 Celcius), and the size of the degree is the same as the Fahrenheit scale, just as 1 Kelvin is the same as 1 Celcius. This means that the conversion between Rankine and Fahrenheit is simple:

$$ T_R = T_F + 459.67 $$

Remember too that the conversion between Celcius and Fahrenheit is given by

$$ T_F = \frac{9}{5}T_C + 32 $$

Together, these give the conversion between Celcius and Rankine.


## Averaging wind direction.

Since the wind direction is given in degrees, we cannot simply average the numerical values of the bearings. Consider the case of 2 values, 1 degree and 359 degrees. We would like this average to 0 (or 360 degrees), i.e. due North, rather than 180 degrees or due South.

One standard way to deal with this is to independently average the sines and cosines of the bearings, and then apply the 2 component $tan^{-1}$ function to convert these back into a bearing. 

$$ a = \frac{\sin(1)+\sin(359)}{2} $$
$$ b = \frac{\cos(1)+\cos(359)}{2} $$

$$ \theta = \tan ^{-1}(a,b) = 0 $$

When performing these calculation, it is important to keep track of units! Computer sin/cos functions tend to work in radians, but for consistency we want to return a value in degrees.

## Formula for dew point

Let's define a function

$$ f(T, r) = \ln\left(\frac{r}{100}\right)+\frac{17.67T}{243.5+T} $$

Then, given air temperature $T$ and relative humidity $r$, we expect the dewpoint $T_d$ to be close to

$$ T_d \approx \frac{243.5f(T, r)}{17.67-f(T, r)} $$


## Types of average

### Arithmetic mean

The arithmetic mean is the sum of a set of values divided by the number of values. This is the most common type of average, and is what most people mean when they say "average".

$ \bar{x} = \frac{1}{n}\sum_{i=1}^{n} x_i $

### Median

The median is the middle value of a set of values, when they are sorted in order. If there is an even number of values, the median is the average of the two middle values.

### Geometric mean

The geometric mean is the $n$th root of the product of $n$ values. This is useful when the values are ratios or rates, as it gives a mean value which is also a ratio or rate.

Equivalently, the geometric mean can be calculated as the exponential of the arithmetic mean of the logarithms of the values.

$ \bar{x} = \left(\prod_{i=1}^{n} x_i\right)^{\frac{1}{n}} $

### Arithmetic-geometric mean

The arithmetic-geometric mean (AGM) is a combination of the arithmetic and geometric means. It is calculated by taking the arithmetic mean of a set of values, and then the geometric mean of the same set of values, and repeating this process until the two means converge to the same value.

As an algorithm, the AGM is calculated as follows:

1. Start with two values, $a_0$ (the arithemtic mean of the data) and $g_0$ (the geometric mean of the data).

2. Calculate the next arithmetic mean $a_{n+1}$ as the average of $a_n$ and $g_n$.

3. Calculate the next geometric mean $g_{n+1}$ as the square root of the product of $a_n$ and $g_n$.

4. Repeat steps 2 and 3 until the two means converge to the same value.

The AGM is useful in many mathematical contexts, and is used in some numerical algorithms.
